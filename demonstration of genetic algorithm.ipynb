```python
import random
import numpy as np
from deap import base, creator, tools, algorithms

#base - It provides the Toolbox class, which acts like a factory where you “register” functions you’ll use in your evolutionary algorithm (e.g., mutation, crossover, selection).
#creator - Used to define new classes (like Individual and Fitness).
#tools - Contains pre-built functions for evolutionary algorithms.Examples:tools.cxTwoPoint → Crossover,tools.mutFlipBit → Mutation,tools.selTournament → Selection


creator.create("FitnessMulti", base.Fitness, weights=(-1.0, 1.0, -1.0))  
creator.create("Individual", list, fitness=creator.FitnessMulti)

NUM_PRODUCTS = 10
NUM_MACHINES = 5
MACHINE_CAPACITIES = [100, 150, 200, 100, 120]  
PRODUCTION_DEADLINES = [20, 15, 25, 18, 22]  
RESOURCE_AVAILABILITY = [10, 15, 12, 8, 10] 

toolbox = base.Toolbox()
toolbox.register("attr_int", random.randint, 0, max(PRODUCTION_DEADLINES))
toolbox.register("individual", tools.initRepeat, creator.Individual, toolbox.attr_int, n=NUM_PRODUCTS * NUM_MACHINES)
toolbox.register("population", tools.initRepeat, list, toolbox.individual)

def evaluate(individual):
    costs = []
    throughputs = []
    energies = []
   
    for i in range(NUM_PRODUCTS):
        costs.append(random.uniform(10, 100))
        throughputs.append(random.uniform(50, 200))
        energies.append(random.uniform(5, 30))
    return sum(throughputs), sum(costs), sum(energies),  

toolbox.register("mate", tools.cxTwoPoint)
toolbox.register("mutate", tools.mutUniformInt, low=0, up=max(PRODUCTION_DEADLINES), indpb=0.05)
toolbox.register("select", tools.selNSGA2)
toolbox.register("evaluate", evaluate)

def main():
    pop = toolbox.population(n=100)
    pop, _ = algorithms.eaMuPlusLambda(pop, toolbox, mu=100, lambda_=100, cxpb=0.7,
                                       mutpb=0.2, ngen=50, stats=None, halloffame=None)
    return pop

if __name__ == "__main__":
    final_population = main()
```
